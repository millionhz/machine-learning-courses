```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(0, 5 ,11)
y = x ** 2
```

# Functional Method

```python
plt.plot(x, y)
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("Test Plot")
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, "r")

plt.subplot(2, 2, 2)
plt.plot(y, x, "b")

plt.subplot(2, 2, 3)
plt.plot(y, x, "g")

plt.show()
```

# Object-Oriented Method

```python
fig = plt.figure()

ax1 = fig.add_axes([1, 1, 1, 1])

ax1.plot(x,y)
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_title("AX1")

plt.show()
```

```python
fig = plt.figure()

fig.suptitle('This is a somewhat long figure title', fontsize=16, x=.9, y=1.95)

ax1 = fig.add_axes([.1, 1, .8, .8])
ax2 = fig.add_axes([.1, .1, .8, .8])
ax3 = fig.add_axes([1, .1, .8, 1.7])

ax1.plot(x, y)
ax1.set_title("AX1")
ax2.plot(y, x)
ax2.set_title("AX2")
ax3.plot(x, y**2)
ax3.set_title("AX3")

plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,8))

ax[0].plot(x, y)
ax[0].plot(x, x**3)

ax[1].plot(y, x)

plt.show()
```

```python
fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(1,1,1)
ax.plot(x, y, color="purple", marker="+", markersize=10)
plt.show()
```
